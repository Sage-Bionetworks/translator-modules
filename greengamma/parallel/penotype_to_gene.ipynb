### Phenotypic Feature to Gene

Consider knowledge pathways from a phenotype to a gene.

```python
from tranql.main import TranQL
t = TranQL ()
```

```python
t("""
set haemoptysis = 'HP:0002105'    
select phenotypic_feature->disease->gene
  from '/graph/gamma/quick'
 where phenotypic_feature = $haemoptysis
   set graph """)
```

```python
t.context.toph (type_name='disease', k='graph', n=500)
```

```python
t.context.toph (type_name='gene', k='graph', n=500)
```

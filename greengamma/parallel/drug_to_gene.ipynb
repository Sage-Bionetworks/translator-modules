## Parallel Implementation Experiments

This notebook looks at the questions, **what drugs target some gene?**
We start with gamma and hope to add other implementations.


```python
from IPython.core.display import display, HTML
from tranql.main import TranQL
tranql = TranQL ()
```

Query a graph for answers.

```python
answer = tranql("""
        set target_gene='HGNC:6871'
        select chemical_substance-[increases_activity_of]->gene
          from '/graph/gamma/quick'
         where gene = $target_gene
           set result """)
```

Have a look at the answer in detail.

```python
display(HTML(answer.toph (type_name='gene', n=50)))
```

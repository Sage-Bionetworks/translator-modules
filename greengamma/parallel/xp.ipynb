## Xeroderma Pigmentosum

Ways of looking at knowledge about XP.


```python
from tranql.main import TranQL
t = TranQL ()
```

```python
t("""
set target_gene='HGNC:2718' -- ddb2 gene
set disease='MONDO:0019600' -- xeroderma pigmentosum

select chemical_substance->gene->disease
  from '/graph/gamma/quick'
 where disease = $disease 
   set treatments""")
```

```python
t.context.toph (type_name='gene', k='treatments', n=500, start=2)
```

```python
t.context.toph (type_name='chemical_substance', k='treatments', n=500, start=200)
```

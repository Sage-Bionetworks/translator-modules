## Clinical Outcome Pathways

Build knowledge networks connecting chemicals to disease through several concepts.

```python
t("""        
set metformin = 'PUBCHEM:4091'
set diabetes_type_2_mellitus = 'MONDO:0015887'

select chemical_substance->gene->biological_process_or_activity->cell->
        anatomical_entity->phenotypic_feature<-disease
  from '/graph/gamma/quick'
 where disease = $diabetes_type_2_mellitus
   and chemical_substance = $metformin
   set graph """)
```

```python
t.context.toph (type_name='anatomical_entity', k='graph', n=500)
```

```python
t("""        
        set asthma = 'MONDO:0004979'
        set salbutamol = 'PUBCHEM:2083'        
        select chemical_substance->gene->biological_process_or_activity->cell->
               anatomical_entity->phenotypic_feature<-disease
          from '/graph/gamma/quick'
         where disease = $asthma
           and chemical_substance = $salbutamol 
           set asthma_graph """)
```

```python
t.context.toph (type_name='biological_process_or_activity', k='asthma_graph', n=500)
```

```python
t("""  
        set asthma = 'MONDO:0004979'
        set imatinib = 'MESH:D000068877' 
        select chemical_substance->gene->biological_process_or_activity->cell->
               anatomical_entity->phenotypic_feature<-disease
          from '/graph/gamma/quick'
         where disease = $asthma
           and chemical_substance = $imatinib 
           set imatinib_graph""")
```

```python
t.context.toph (type_name='chemical_substance', k='imatinib_graph', n=500)
```

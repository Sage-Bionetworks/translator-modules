## MHC processes

Major histocompatibility complex (MHC) refers to a large group of genes that code for proteins that play an essential role in which of the following?

```python
#Load some functions for calling the quick service and parsing its output
import pandas as pd
import os
import sys
from IPython.core.display import display, HTML

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from gg_functions import parse_answer, quick
```

```python
def family_process(family_id):
    return {
    "machine_question": {
        "nodes": [
            {
                "id": "n0",
                "type": "gene_family",
                "curie": family_id
            },
            {
                "id": "n1",
                "type": "gene",
                "set": True
            },
            {
                "id": "n2",
                "type": "biological_process_or_activity"
            }
        ],
        "edges": [
            {
                "id": "e0",
                "source_id": "n0",
                "target_id": "n1"
            },
            {
                "id": "e1",
                "source_id": "n1",
                "target_id": "n2"
            }
        ]
    }
}
```

```python
q = family_process('HGNC.FAMILY:588')
a = quick(q)
df = parse_answer(a,node_list=['n2'])
```

```python
with pd.option_context('display.max_colwidth', 500):
    display(df)
```

The options are:
* Antigen presentation to T lymphocytes
* Phagocytosis by neutrophils
* Antigen presentation to B lymphocytes
* Phagocytosis by macrophages


Between rows 3-5, we infer that the answer is `Antigen presentation to T lymphocytes`

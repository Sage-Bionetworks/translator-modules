```python
import requests
```

```python
headers = {'content-type': 'application/x-www-form-urlencoded'}
```

```python
# replace "MONDO:0016575, MONDO:0009101" with list of MONDO IDs of your interest
params = 'ids=MONDO:0016575, MONDO:0009101&fields=disgenet.genes_related_to_disease'
```

```python
res = requests.post('http://mydisease.info/v1/disease', data=params, headers=headers)
res.json()
```
